```python
import numpy as np
import urllib
import bs4
import pandas as pd

from bs4 import BeautifulSoup
from urllib import request
```

```python

#******************Webscraping*******************
#pour les premières pages, le tableau voulu est le 6e
nbre_pages=np.array([23,24,25,26,53]) 

data_medailles = dict()
for i in nbre_pages: 
    request_text_i = request.urlopen("http://www.olympedia.org/editions/" + str(i))
    pagei=bs4.BeautifulSoup(request_text_i,"lxml")
    tableau_medailles = pagei.find_all("table")[5]
    rows = tableau_medailles.find_all("tr")
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        #print(cols)

    dico_medailles = dict()
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        if len(cols) > 0 : 
            dico_medailles[cols[0]] = cols[1:]

   # print(dico_medailles)
    data_medailles[i] = pd.DataFrame.from_dict(dico_medailles,orient='index')
    #data_medailles.columns = columns_medailles[0:]
    #print(data_medailles)
#data_medailles


    
#pour les dernières pages, le tableau voulu est le 5e

nbre_pages=np.array([54,59,61]) 

for i in nbre_pages: 
    request_text_i = request.urlopen("http://www.olympedia.org/editions/" + str(i))
    pagei=bs4.BeautifulSoup(request_text_i,"lxml")
    tableau_medailles = pagei.find_all("table")[4]
    rows = tableau_medailles.find_all("tr")
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        #print(cols)

    dico_medailles = dict()
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        if len(cols) > 0 : 
            dico_medailles[cols[0]] = cols[1:]

    data_medailles[i] = pd.DataFrame.from_dict(dico_medailles,orient='index')
    #data_medailles = pd.DataFrame.from_dict(dico_medailles,orient='index')
    #data_medailles.columns = columns_medailles[0:]
    #print(data_medailles)

for i in np.array([23,24,25,26,53,54,59,61]):
    data_medailles[i].columns = ['Code','Gold','Silver','Bronze','Total']
data_medailles



```

```python
#Ajout des années et lieux

data_medailles[23] = pd.DataFrame.assign(data_medailles[23],Annee=1992,Lieu="Barcelona")
data_medailles[24] = pd.DataFrame.assign(data_medailles[24],Annee=1996,Lieu="Atlanta")
data_medailles[25] = pd.DataFrame.assign(data_medailles[25],Annee=2000,Lieu="Sydney")
data_medailles[26] = pd.DataFrame.assign(data_medailles[26],Annee=2004,Lieu="Athina")
data_medailles[53] = pd.DataFrame.assign(data_medailles[53],Annee=2008,Lieu="Beijing")
data_medailles[54] = pd.DataFrame.assign(data_medailles[54],Annee=2012,Lieu="London")
data_medailles[59] = pd.DataFrame.assign(data_medailles[59],Annee=2016,Lieu="Rio de Janeiro")
data_medailles[61] = pd.DataFrame.assign(data_medailles[61],Annee=2020,Lieu="Tokyo")
data_medailles
  
```
