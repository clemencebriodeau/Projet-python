Sur https://fr.wikipedia.org/wiki/Liste_des_codes_pays_du_CIO#cite_note-ND-3 
on apprend que les codes pays du CIO (Comité international olympique) sont différents des codes ISO 3166-1 utilisés par la Banque mondiale.

Donc on fait du webscrapping pour récupérer le tableau permettant de lier les codes du CIO aux codes de la Banque mondiale.

```python
#Récupération du tableau de conversion des codes pays sur wikipedia pour avoir les codes de la banque mondiale

urlpage = 'https://fr.wikipedia.org/wiki/Liste_des_codes_pays_du_CIO?fbclid=IwAR2JKLA2BJI_LIWsV-Zd5n437rFtOEkRYWpqFUTi9qo5WyHx7OVxioiGmi8'

#renvoie le html à la variable request_text
request_text = request.urlopen(urlpage).read()

page1 = bs4.BeautifulSoup(request_text, "lxml")
#print(page1.find("table"))

tableau_codes = page1.find_all("table")[1]
#print(tableau_codes)


# on recherche toutes les lignes du tableau avec la balise "tr"
rows = tableau_codes.find_all("tr")


for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    #print(cols)

dico_codes = dict()
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 0 : 
        dico_codes[cols[0]] = cols[1:]


df_codes = pd.DataFrame.from_dict(dico_codes,orient='index')
df_codes.columns=['Pays','ISO 3166-1']
df_codes
```

```python
df_codes.reset_index(inplace=True) #la colonne des codes du CIO était utilisée comme index, donc on réinitialise l'index
df_codes.rename({"index": "Code CIO"}, axis=1, inplace=True)
df_codes
```
