## Investissement dans le sport


Il n’existe aucune base de données recensant directement l'investissement public dans le sport pour l’ensemble des pays. Mais l'OCDE a développé un système de classification des dépenses gouvernementales : COFOG (Classification of Functions of the Government). Pour approximer l'investissement public dans le sport, on retient la COFOG "recreation, culture and religion". On utilise les données du FMI et on complète pour avoir moins de valeurs manquantes avec celles d'Eurostat.

Données FMI téléchargées sur le lien suivant : https://data.imf.org/?sk=5804c5e1-0502-4672-bdcd-671bcdc565a9

Données Eurostat téléchargées sur le lien suivant : https://ec.europa.eu/eurostat/databrowser/view/GOV_10A_EXP__custom_1784023/default/table?lang=en

```python
invest_sport_FMI = pd.read_excel("Invest_sport_recreation_culture_religion_FMI_domesticcurrency.xlsx", header=2)
```

```python
invest_sport_FMI.rename({"Unnamed: 0": "Country name"}, axis=1, inplace=True)
#on renomme la colonne des pays

invest_sport_FMI.head(10)
```

```python
invest_sport_Eurostat = pd.read_excel("Invest_sport_recreation_culture_religion_Eurostat_domesticcurrency.xlsx", 
                                      sheet_name="Sheet 1", header=10)
```

```python
invest_sport_Eurostat.dropna(axis=1, how='all',inplace=True)
#on supprime les colonnes qui ne continennent aucune valeur, entre chaque année

invest_sport_Eurostat.rename({"TIME": "Country code", "TIME.1": "Country name"}, axis=1, inplace=True)
#on renomme les deux premières variables

invest_sport_Eurostat.drop(0, axis=0, inplace=True)
#on supprime la 1ère ligne qui contient juste les anciens noms des deux premières variables

invest_sport_Eurostat.replace([":"], np.nan, inplace=True)
#les valeurs manquantes étaient représentées par :. On les remplace par NaN
```
